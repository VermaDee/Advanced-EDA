```python
df2=df1.copy()
df2.head()
```

```python
df2= df2.dropna(thresh=df2.shape[1] * 0.80, axis=0).reset_index(drop=True)
df2.shape
```

```python
df2.head()
```

```python
df2.isna().mean().sort_values().plot(
    kind="bar", figsize=(15, 4),
    title="Percentage of missing values per feature",
    ylabel="Ratio of missing values per feature");
```

```python
df2 = df2.dropna(thresh=df2.shape[0] * 0.85, axis=1)
df2.shape
```

```python
plt.figure(figsize=(10, 6))#Scatter plot
ax = sns.scatterplot(x ='work_rate',y = df2['wage_eur'],hue = "league_rank",data = df2,
                     palette = ["green", "red", "coral", "blue"],legend="full",alpha = 0.4)

#Getting the max wage for each work rate
max_wage_eur = df2.groupby("work_rate")["wage_eur"].max() #Making a line plot of max wages
sns.lineplot(data = max_wage_eur,ax = ax.axes,color="grey"),ax.tick_params(axis= "x", rotation=90)
plt.xlabel("Work Rate")
plt.ylabel("Wage EUR")
plt.title("Relationship between work rate and wage by league rank", fontsize = 18)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 6), dpi=80)
sns.stripplot(x = "nationality", y = "wage_eur",
              data=df_clubs, size = 7, ax=ax)
plt.tick_params(axis='x', which='major', labelsize=12, rotation=90)
plt.xlabel("Nationality")
plt.ylabel("Wage")
plt.title("Wages for Nationalities in Club Level")
plt.show()
```

```python
plt.style.use("default")
plt.figure(figsize=(5,5))
plt.scatter(df2.age, df2.wage_eur, 
            alpha = 0.8, c = df2.wage_eur, 
            cmap = "Spectral")

plt.colorbar(orientation = "vertical")
plt.xlabel("Age")
plt.ylabel("Wage")
plt.title("Age vs Wage")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.style.use("default")
plt.figure(figsize=(5,5))
plt.scatter(df2.height_cm, df2.weight_kg,
            alpha = 0.8, c = df2.weight_kg, 
            cmap = "Spectral")

plt.colorbar(orientation = "vertical")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")
plt.grid(True)
plt.legend()
plt.show()
```
